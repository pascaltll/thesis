```python
import re
from collections import Counter

def cargar_datos(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        textos = f.readlines()
    return [texto.strip() for texto in textos]

def limpiar_texto(texto, umbral):
    palabras = re.findall(r'\b\w+\b', ' '.join(texto).lower())  # Unir todo el texto para contar frecuencias
    frecuencia = Counter(palabras)
    
    texto_limpio = []
    for linea in texto:
        # Tokenización y filtrado de palabras por línea
        #palabras_linea = re.findall(r'\b\w+\b', linea.lower())
        palabras_filtradas = [palabra for palabra in texto if frecuencia[palabra] <= umbral]
        
        # Reconstruir cada línea con las palabras filtradas
        texto_limpio.append(' '.join(palabras_filtradas))
    
    # Unir las líneas filtradas con saltos de línea
    return '\n'.join(texto_limpio)
```

```python
from nltk.tokenize import word_tokenize
import nltk
from nltk.probability import FreqDist
from tqdm import tqdm
nltk.download('all')
def cargar_datos(archivo):
    with open(archivo, 'r', encoding='utf-8') as f:
        textos = f.readlines()
    return [texto.strip() for texto in textos]

def guardar_datos(dir_arichivo, texto):
    with open(dir_arichivo, "w", encoding="utf-8") as f:
        for linea in texto:
            f.write(linea + "\n")
        
def procesar_frecuencia(path_original,umbral_frecuencia,path_save):
    texto = cargar_datos(path_original)

    fdist = FreqDist(word_tokenize(' '.join(texto)))
    stop_words = set(word for word, count in fdist.items() if count >= umbral_frecuencia)
    
    #separar por generos
    nuevo_texto= []
    for fragmento in tqdm(texto, desc="tqlm"):  
        palabras = word_tokenize(fragmento)
        palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]
        nuevo_texto.append(' '.join(palabras_filtradas))
        
    #gaurdar los datatos
    guardar_datos(path_save, nuevo_texto)
    
    print("tqlm: Procesamiento completado.")



    
```

```python
path_original1 = '../dataset/Первый_жанр_исходная.txt' 
path_original2 = '../dataset/Второй_жанр_исходная.txt' 
path_save_1 = "../dataset/сокращение по частотности/1е_Изъяты лексемы с частотой выше 5.txt"
path_save_2 = "../dataset/сокращение по частотности/2е_Изъяты лексемы с частотой выше 5.txt"
umbral_frecuencia = 5
procesar_frecuencia(path_original1, umbral_frecuencia, path_save_1)
procesar_frecuencia(path_original2, umbral_frecuencia, path_save_2)

           
```
